```python
import pandas as pd
import requests
import numpy as np

# 1) seu código aqui
sinasc = pd.read_csv('SINASC_RO_2019.csv')
print(sinasc.shape)
sinasc.drop_duplicates().shape
# Não há duplicados
```

```python
# 2) seu código aqui
missing_counts = sinasc.isna().sum()

print("A contagem de valores missing do Dataframe é:")
print(missing_counts)
```

```python
# 3) seu código aqui
colunas_interesse = ['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO',
                     'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5']

sinasc_1 = sinasc[colunas_interesse]
sinasc_1.head()
```

```python
missing_counts_1 = sinasc_1.isnull().sum()

print("A contagem de valores missing do Dataframe é:")
print(missing_counts_1)
```

```python
# 4) seu código aqui
numero_linhas = len(sinasc_1['APGAR5'])
print("O número de linhas de 'APGAR5' é:", numero_linhas)

missing_counts_2 = sinasc_1['APGAR5'].isnull().sum()
print("O número de missings é:", missing_counts_2)
```

```python
sinasc_1 = sinasc_1.dropna(subset = ['APGAR5'])
sinasc_1
```

```python
# 5) seu código aqui

sinasc_1.loc[:, 'ESTCIVMAE'] = sinasc_1['ESTCIVMAE'].replace(9, 'ignorado')
sinasc_1.loc[:, 'CONSULTAS'] = sinasc_1['CONSULTAS'].replace(9, 'ignorado')

sinasc_1
```

```python
# 6) Seu código aqui
sinasc_1.loc[:,'QTDFILVIVO'] = sinasc_1['QTDFILVIVO'].fillna(0)

sinasc_1['QTDFILVIVO']
```

```python
# 7) seu código aqui
missing_counts_3 = sinasc_1.isnull().sum()

print("O número de dados faltantes nesse dataframe é:")
print(missing_counts_3)
```

```python
# Poderemos preencher a variável que mostra o estado civil da mãe pelo
# valor anterior, por não ser tão relevante na situação do recém-nascido

sinasc_1['ESTCIVMAE'] = sinasc_1['ESTCIVMAE'].fillna(method = 'bfill')

# Também podemos preencher a variável que mostra a escolaridade da mãe
# pois no momento do nascimento e na primeira respiração, essa variável
# não é determinante para a saúde do bebê

sinasc_1['ESCMAE'] = sinasc_1['ESCMAE'].fillna(method = 'bfill')
```

```python
# 8) seu código aqui

sinasc_1.loc[sinasc_1['APGAR5'] < 4, 'ASFIXIA'] = 'asfixia severa'

sinasc_1.loc[(sinasc_1['APGAR5'] >= 4) & 
             (sinasc_1['APGAR5'] <= 5), 'ASFIXIA'] = 'asfixia moderada'

sinasc_1.loc[(sinasc_1['APGAR5'] >= 6) & 
             (sinasc_1['APGAR5'] <= 7), 'ASFIXIA'] = 'asfixia leve'

sinasc_1.loc[sinasc_1['APGAR5'] > 7, 'ASFIXIA'] = 'normal'

sinasc_1
```

```python
frequencias = sinasc_1['ASFIXIA'].value_counts()

print("A frequência de cada valor na coluna ASFIXIA é:")
print(frequencias)
```
